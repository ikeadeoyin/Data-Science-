```python
#split out validation test_data
array = data.values
X = array[:,0:4]
y = array[:,4]
```

```python
X
```

```python
#split data into training set and testing set
X_train, X_test, y_train, y_test = train_test_split(
  X, y, test_size=0.25, random_state=0) #test size = 25% of the dataset.
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```python
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
#create dataframe for X_train
#label the columns usings the strings in data
train_dataframe = pd.DataFrame(X_train, columns = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"])
train_dataframe
```

```python
sns.pairplot(train_dataframe)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn =  KNeighborsClassifier(n_neighbors=1)
```

```python
#to build the model on the training set
knn.fit(X_train, y_train)
```

```python
#make predictions
y_pred = knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred))
#print("Predicted target name: {}".format)
```
